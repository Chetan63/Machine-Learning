# TASK 1: Prediction using Supervised Machine Learning

Predict the percentage of a student based on the number of study hours

# Step 1:Importing the libraries

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Step 2: Importing the data from the given link

```python
# import data
data=pd.read_csv('http://bit.ly/w-data')
```

```python
# view data (first five rows)
data.head()
```

```python
# view data (last five rows)
data.tail()
```

# Step 3: Visualizing the data

```python
# visualize data
data.plot(x='Hours',y='Scores',style='o')
plt.title('hours vs scores')
plt.xlabel('hours of study')
plt.ylabel('scores in percentage')
plt.grid(True)
plt.show()
```

# Step 4: Data Spilitting

```python
# divide the data into attributes and labels
X=data.iloc[:, :-1].values
Y=data.iloc[:, 1].values
```

```python
# now we have to split the data into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.33,random_state=2529)
```

# Step 5: Training the linear ragression model

```python
# training the algorithm
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train, Y_train)
```

# Step 6: Plotting the regression line

```python
# plotting the regression line
line= model.coef_*X + model.intercept_
plt.scatter(X,Y)
plt.plot(X, line, color='red')
plt.show()
```

# Step 7: Making Predictions

```python
# predictions with our trained model
pred_score=model.predict(X_test)
print(pred_score)
```

# Step 8: Comparing actual with predicted output

```python
# comparing Actual with predicted
df=pd.DataFrame({'Actual':Y_test, 'Predicted':pred_score})
df
```

```python
hours=[[9.25]]
score_pred=model.predict(hours)
print("number of hours(study): {}".format(hours))
print("predicted score(percentage): {}".format(score_pred[0]))
```

# Step 9: Evaluating the performance of the model

```python
# evaluating the performance of the model
from sklearn import metrics
print('Mean Absolute Error:',metrics.mean_absolute_error(Y_test,pred_score))
```
